### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
colormap = cm.get_cmap('YlOrRd') # Yellow-Orange-Red colormap

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total por cada tienda
# Definir lista de tiendas y ingresos total
tiendas = {'Tienda': tienda, 'Tienda2': tienda2, 'Tienda3': tienda3, 'Tienda4': tienda4}
Ingresos_tiendas = {}
for nombre_tienda, id_tienda in tiendas.items():
  Ingresos_tiendas[nombre_tienda] = sum(id_tienda['Precio'])

display(Ingresos_tiendas)


nombres_tiendas_bar = list(Ingresos_tiendas.keys())
ingresos_totales_bar = list(Ingresos_tiendas.values())


colormapv = cm.get_cmap('viridis', len(nombres_tiendas_bar))
coloresv = [colormapv(i) for i in range(len(nombres_tiendas_bar))]


plt.figure(figsize=(10, 6))
plt.barh(nombres_tiendas_bar, ingresos_totales_bar, color=coloresv)
plt.xlabel('Ingresos Totales')
plt.ylabel('Tienda')
plt.title('Ingresos Totales por Tienda')
plt.grid(axis='x')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
#conseguir todos el promedio de clasificacion
calificacion_promedio_tiendas = {}

for nombre_tienda, tienda in tiendas.items():
  total_calificacion = 0
  cantidad_calificaciones = 0
  for calificacion in tienda['Calificación']:
    total_calificacion += calificacion
    cantidad_calificaciones += 1
  calificacion_promedio = total_calificacion / cantidad_calificaciones
#redondear con 1 decimal
  calificacion_promedio_r = round(calificacion_promedio + 0.05, 1)
  calificacion_promedio_tiendas[nombre_tienda] = calificacion_promedio_r

display("El promedio de calificación para cada tienda es:")
display(calificacion_promedio_tiendas)
```

# 4. Productos más y menos vendidos

```python
# Tienda
# Contar las ventas de cada producto
conteo_productos_tienda = {}
for producto in tienda['Producto']:
  if producto in conteo_productos_tienda:
    conteo_productos_tienda[producto] += 1
  else:
    conteo_productos_tienda[producto] = 1

# Encontrar el producto más vendido
producto_mas_vendido_tienda = max(conteo_productos_tienda, key=conteo_productos_tienda.get)
cantidad_mas_vendida_tienda = conteo_productos_tienda[producto_mas_vendido_tienda]

# Encontrar el producto menos vendido
producto_menos_vendido_tienda = min(conteo_productos_tienda, key=conteo_productos_tienda.get)
cantidad_menos_vendida_tienda = conteo_productos_tienda[producto_menos_vendido_tienda]

# Mostrar los resultados
display(f"En Tienda:")
display(f"El producto más vendido es '{producto_mas_vendido_tienda}' con {cantidad_mas_vendida_tienda} ventas.")
display(f"El producto menos vendido es '{producto_menos_vendido_tienda}' con {cantidad_menos_vendida_tienda} ventas.")


tienda_conteo = pd.DataFrame(list(conteo_productos_tienda.items()), columns=['Producto', 'Cantidad'])
tienda_conteo = tienda_conteo.sort_values(by='Cantidad', ascending=False)

productos_ordenados = tienda_conteo['Producto']
cantidades_ordenadas = tienda_conteo['Cantidad']



colores = colormap(cantidades_ordenadas / max(cantidades_ordenadas)) # Mapear cantidades a colores

plt.figure(figsize=(15, 7))
plt.bar(productos_ordenados, cantidades_ordenadas, color=colores)
plt.ylabel('Cantidad de Ventas')
plt.title('Cantidad de Ventas por Producto en Tienda (Mayor a Menor)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()



```

# 5. Envío promedio por tienda

```python
#Promedio costo de envío
costo_envio_promedio_tiendas = {}

for nombre_tienda, cada_tienda in tiendas.items():
  total_costo_envio = 0
  cantidad_envios = 0
  for costo in cada_tienda['Costo de envío']:
    total_costo_envio += costo
    cantidad_envios += 1
  costo_envio_promedio = total_costo_envio / cantidad_envios

  costo_envio_promedio_redondeado = round(costo_envio_promedio, 2)
  costo_envio_promedio_tiendas[nombre_tienda] = costo_envio_promedio_redondeado

display("El costo de envío promedio para cada tienda es:")
display(costo_envio_promedio_tiendas)

# Costo total de envío por cada tienda
Costos_envio_tiendas = {}
for nombre_tienda, id_tienda in tiendas.items():
  Costos_envio_tiendas[nombre_tienda] = sum(id_tienda['Costo de envío'])

display("Costo total de envío por tienda:")
display(Costos_envio_tiendas)
```
